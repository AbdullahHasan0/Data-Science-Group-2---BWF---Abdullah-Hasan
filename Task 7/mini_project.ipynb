# Car Rental System Using OOP 
The Vehicle Rental System is a simplified simulation of a rental service that allows customers to rent and return vehicles, as well as display available vehicles.

# Coffee Machine
This Program simulate the coffee machine which can make 3 types of coffees which are cappuccino , latte and espresso. Each type of coffee required different amount of resources and each have different price. In this program there are methods for refilling resources (milk, coffee ,water).Also method for counting and return change amount to customer. 

```python
class CoffeeType:
    def __init__(self, name, water, milk, coffee, cost):
        self.name = name
        self.water = water
        self.milk = milk
        self.coffee = coffee
        self.cost = cost

class CoffeeMachine:
    def __init__(self, water, milk, coffee):
        self.water = water
        self.milk = milk
        self.coffee = coffee
        self.cash = 0

    def report(self):
        print(f"Remaining Water: {self.water} ml")
        print(f"Remaining Milk : {self.milk} ml")
        print(f"Remaining Coffee : {self.coffee} gm")
        print(f"Cash Collected : {self.cash} $")

    # def addNewCoffeeType(self):
    #     name = input("Enter name of coffee type: ")
    #     water = int(input("Enter amount of water required: "))
    #     milk = int(input("Enter amount of milk required: "))
    #     coffee = int(input("Enter amount of coffee required: "))
    #     cost = int(input("Enter cost of coffee: "))
    #     coffee_type = CoffeeType(name, water, milk, coffee, cost)
        # self.coffee_types.append(coffee_type)

    def is_resource_sufficient(self, coffee_type):
        if self.water < coffee_type.water:
            print("Not enough water.")
            return False
        if self.milk < coffee_type.milk:
            print("Not enough milk.")
            return False
        if self.coffee < coffee_type.coffee:
            print("Not enough coffee.")
            return False
        return True

    def process_coins(self):
        print("Please insert coins.")
        quarters = int(input("Enter amount of quarters: "))
        dimes = int(input("Enter amount of dimes: "))
        nickels = int(input("Enter amount of nickels: "))
        pennies = int(input("Enter amount of pennies: "))
        total = (quarters * 0.25) + (dimes * 0.10) + (nickels * 0.05) + (pennies * 0.01)
        return total

    def make_coffee(self, coffee_type, cash):
        if not self.is_resource_sufficient(coffee_type):
            print("Not enough resources. Money refunded.")
            return False
        if cash >= coffee_type.cost:
            change = cash - coffee_type.cost
            self.water -= coffee_type.water
            self.milk -= coffee_type.milk
            self.coffee -= coffee_type.coffee
            self.cash += coffee_type.cost
            print(f"Here is your {coffee_type.name}. Enjoy!")
            print(f"Here is your change: ${change:.2f}")
            return True
        else:
            print("Not enough money. Money refunded.")
            print(f"Required ${coffee_type.cost:.2f}, you provided ${cash:.2f}")
            return False

    def run(self):
        coffeeTypes = {
            "espresso": CoffeeType("Espresso", 50, 0, 18, 1.50),
            "latte": CoffeeType("Latte", 200, 150, 24, 2.50),
            "cappuccino": CoffeeType("Cappuccino", 250, 100, 24, 3.00)
        }

        while True:
            print("1. Espresso")
            print("2. Latte")
            print("3. Cappuccino")
            print("4. Report")
            # print("5. Add New Coffee Type")
            print("5. Exit")
            choice = input("What would you like?:\n").lower()
            if (choice == "1"):
                choice = "espresso"
            elif (choice == "2"):
                choice = "latte"
            elif (choice == "3"):
                choice = "cappuccino"
            if choice == "4":
                self.report()
            # elif choice == "5":
            #     self.addNewCoffeeType()
            elif choice == "5":
                break
            elif choice in coffeeTypes:
                coffee_type = coffeeTypes[choice]
                cash = self.process_coins()
                self.make_coffee(coffee_type, cash)
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    coffee_machine = CoffeeMachine(1000, 1000, 1000)
    coffee_machine.run()
```
