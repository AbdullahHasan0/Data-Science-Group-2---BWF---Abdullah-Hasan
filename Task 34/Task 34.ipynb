#### What is Deep Learning ?

Deep learning is a subset of machine learning that utilizes multiple layers of artificial neurons to model complex patterns in data. It aims to learn representations of the input data that can be used for various tasks, such as classification, regression, and recommendation systems.

#### Importance of Deep Learning in AI

Deep learning plays a crucial role in various AI applications, including image recognition, natural language processing, speech recognition, and game playing. By leveraging the power of neural networks, deep learning algorithms can learn complex patterns and representations from large amounts of data, enabling more accurate and efficient decision-making in various domains.

#### Types of Deep Learning

There are various types of deep learning architectures, such as convolutional neural networks (CNNs), recurrent neural networks (RNNs), and long short-term memory (LSTM) networks. Each type has its own advantages and limitations, and understanding their underlying mechanisms is essential for designing effective deep learning models.

#### Advantages of Deep Learning

- Improved performance: Deep learning algorithms have shown significant improvements in various tasks, such as image recognition, natural language processing, and speech recognition.
- Efficient training: Deep learning models can train on large datasets more efficiently than traditional machine learning algorithms.
- Flexibility: Deep learning models can capture complex patterns and representations, making them suitable for a wide range of applications.
- Robustness: Deep learning models can handle noise and outliers effectively, making them more reliable for real-world scenarios.

#### Limitations of Deep Learning

- Computational resources: Deep learning models require significant computational resources, such as GPU acceleration and parallel processing capabilities. Training large-scale deep learning models can be time-consuming and resource-intensive.
- Overfitting: Deep learning models can be prone to overfitting, especially when trained on small datasets. Regularization techniques, such as dropout and data augmentation, can help prevent overfitting.
- Interpretability: Deep learning models can be difficult to interpret, making it challenging to understand their decision-making processes.

#### Deep Learning Frameworks and Libraries

There are several popular deep learning frameworks and libraries available, such as TensorFlow, PyTorch, and Keras. TensorFlow is an open-source library for machine learning developed by Google. PyTorch is a popular deep learning framework developed by Facebook. Keras is a high-level API for TensorFlow, providing a user-friendly interface for building and training deep learning models.


#### What is Neural Network ?

Neural networks are computational models inspired by the structure and function of the human brain. They consist of interconnected nodes, or neurons, arranged in layers to process and learn patterns in data. Each neuron receives input from other neurons, processes it, and produces an output. Neural networks have been used in various applications, such as image recognition, natural language processing, speech recognition, and game playing.

### Splitting Data into Training and Testing Data

```python
X = data.drop('Sales',axis=1)
y = data["Sales"]

X_train , X_test , y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=40)
print(X_train.shape)

print(y_train.shape)
print(X_train)

```

### Making Linear Regression Model For Comparision

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

### Building a ANN

```python
model = keras.Sequential()

# Add the hidden layer
model.add(Dense(units=10, input_shape=(3,), activation='relu'))

# Add the output layer
model.add(Dense(units=1))

model.summary()
```

### Model Compilation

#### What is an Optimizer?

An optimizer is a critical component in machine learning and deep learning models. Its primary function is to minimize the loss function or maximize the performance metric of a model by adjusting the model's parameters in an iterative process.

#### How does an Optimizer work?

Here's a high-level overview of the optimization process:

***Forward Pass:***  
The model processes the input data and produces an output.

***Loss Calculation:***
 The difference between the predicted output and the actual output is calculated using a loss function (e.g., mean squared error, cross-entropy).

***Backward Pass:*** 
 The gradients of the loss with respect to the model's parameters are computed.

***Optimizer Update:*** 
 The optimizer updates the model's parameters based on the gradients and the optimization algorithm used.
Types of Optimizers

Some popular optimizers include:

***Stochastic Gradient Descent (SGD):***
 Updates parameters based on a single sample from the training dataset.

***Momentum:***
 Adds a momentum term to the update rule, helping the optimizer escape local minima.

***Nesterov Accelerated Gradient:***
 A variant of momentum that incorporates a different update rule.
Adagrad: Adaptive learning rate optimizer that adjusts the learning rate for each parameter based on the gradient norm.

***Adam:*** 
Adaptive learning rate optimizer that combines the benefits of Adagrad and RMSProp.

***RMSProp:***
Divides the learning rate by an exponentially decaying average of squared gradients.

***Conjugate Gradient:***
 An optimization algorithm that uses a sequence of conjugate directions to minimize the loss function.

 ### Loss Function in TensorFlow ?

In TensorFlow, a loss function is used to measure the difference between the model's predictions and the actual labels. The goal of training a model is to minimize the loss function.

### Types of Loss Functions
1. ***Mean Squared Error (MSE)*** :
Used for regression problems
Measures the average squared difference between predictions and actual values

2. ***Cross-Entropy Loss***: 
Used for classification problems
Measures the difference between the predicted probabilities and the true labels

3. ***Binary Cross-Entropy Loss***:
Used for binary classification problems
Measures the difference between the predicted probabilities and the true labels

4. ***Mean Absolute Error (MAE)***:
Used for regression problems
Measures the average absolute difference between predictions and actual values

5. ***Hinge Loss***:
Used for binary classification problems
Measures the difference between the predicted scores and the true labels



```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=  .001),loss= 'mse')
```

###  Model Training

```python
history = model.fit(X_train,y_train,batch_size=16,epochs=2000,validation_split=0.03)
```

### Plotting The Training Results

```python
plt.plot(history.history['loss'],'g',label = 'Training Loss')

plt.plot(history.history['val_loss'],'b',label = 'Validation Loss')

plt.title('Model Loss')

plt.xlabel('Epochs')

plt.ylabel('Loss')

plt.legend()

plt.show()
```

### Prediction By Model

```python
print('MSE:',model.evaluate(X_test, y_test))
```

```python
y_pred = model.predict(X_test)

### Comparing Predictions with Actual Values

plt.scatter(y_test, y_pred)

plt.title('Predicted vs Actual Values')

plt.xlabel('Actual Values')

plt.ylabel('Predicted Values')

plt.show()

```

### Residual Plot

```python
y_pred = model.predict(X_test).flatten()

residuals = y_test - y_pred

plt.scatter(y_test,residuals)

plt.title('Residual Plot')

plt.xlabel('True Values')

plt.ylabel('Residuals')

plt.show()

```

### Interpretation 

The residual plot shows a scatter plot of the true values against the residuals, which indicates that the model's predictions are relatively consistent with the actual labels.
