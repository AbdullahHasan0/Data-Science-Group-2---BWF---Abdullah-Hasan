<h3>Customize the subplots with titles, axis labels, and legends.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'weather' is your DataFrame

# Adjusting the figure size
plt.figure(figsize=(12, 12))

# plot 1 (Line Plot)
x = weather.index[:11]
y = weather["Temperature"][:11]
plt.subplot(2, 2, 1)
plt.plot(x, y, label="Temperature Over Time", color='b')
plt.title("Temperature Over Time")
plt.xlabel("Index")
plt.ylabel("Temperature")
plt.legend()

# plot 2 (Scatter Plot)
x = weather["Temperature"][:100]
y = weather["UV Index"][:100]
plt.subplot(2, 2, 2)
plt.scatter(x, y, label="Temperature vs UV Index", color='r')
plt.title("Temperature vs UV Index")
plt.xlabel("Temperature")
plt.ylabel("UV Index")
plt.legend()

# plot 3 (Bar Plot)
seasonInfo = weather.groupby("Season")["Temperature"].mean()
x = seasonInfo.index
y = seasonInfo.values
plt.subplot(2, 2, 3)
plt.bar(x, y, color=['green', 'yellow', 'orange', 'skyblue'],label = ["Winter","Spring","Summer","Autumn"])
plt.title("Average Temperature per Season")
plt.xlabel("Season")
plt.ylabel("Temperature")

plt.legend()

# plot 4 (Histogram Plot)
plt.subplot(2, 2, 4)
plt.hist(weather["Temperature"], bins=15, color="pink", alpha=0.7, histtype="barstacked", rwidth=0.7, label="Temperature Distribution")
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

<h2>Box Plots

<h3> 1. Create a box plot for a given numerical data set.

```python
plt.boxplot(weather["UV Index"])
```

<h3>2. Generate multiple box plots in a single figure to compare different data sets.


```python
plt.boxplot(weather["UV Index"],positions=[1],widths=0.6,label="UV Index")
plt.boxplot(weather["Temperature"],positions=[2],widths=0.6,label="Temperature")
plt.xticks([1,2],["UV Index", "Temperature"])
plt.legend()

plt.show()
```

<h3>3. Customize the box plots with titles, labels, and change the colors.

Some Useful Attributes for Customizing Box Plot

- notch: *Notched box plots apply a "notch" or narrowing of the box around the median.*
- patch_artist: *This attribute allows you to customize the appearance of the box plot boxes.*
- boxprops: *This attribute allows you to customize the appearance of the box itself.*
- medianprops: *This attribute allows you to customize the appearance of the median lines.*
- whiskerprops: *This attribute allows you to customize the appearance of the whiskers.*
- flierprops: *This attribute allows you to customize the appearance of the outliers.*
- capprops: *This attribute allows you to customize the appearance of the caps on the whiskers.*


```python
boxPlotData = [weather["UV Index"],weather["Temperature"]]
plt.boxplot(boxPlotData,widths=0.6,labels=["UV Index","Temparature"],notch='True',patch_artist = True,boxprops=dict(color='green',facecolor= "pink"),medianprops=dict(color='red'),whiskerprops=dict(color='purple'),flierprops=dict(markerfacecolor = "yellow", marker="s"),)

plt.title("UV Index and Temperature Distribution")
plt.xlabel("Parameters")
plt.ylabel("Values")


plt.legend(["UV Index","Temparature"])
plt.show()
```

<h2> Stem Plot </h2>

What is Stem Plot ? 

A stem plot is a type of plot that shows data points on a vertical line. Each data point is represented by a marker (the "head") and a line extending from a baseline to the marker (the "stem"). Stem plots are useful for visualizing discrete data points and their distribution along an axis. They are particularly helpful when you want to emphasize individual values and their positions.

```python
x = weather.index[:51]
y = weather.Temperature[:51]



plt.title("Temperature Over Time")
plt.xlabel("Index")
plt.ylabel("Temperature")

plt.stem(x, y, linefmt='g', markerfmt='bD')
```

<h2> Stack Plot </h2>

What is Stack Plot?

A stack plot is a type of graph that is used to display the cumulative results of multiple data series over time or another continuous variable. It is useful for showing how different components contribute to the whole, making it easy to visualize changes in proportions.

```python

x = weather.index[:101]
temperature=weather["Temperature"][:101]
uvIndex=weather["UV Index"][:101]
precipitation=weather["Precipitation (%)"][:101]

# Stack plot
color = ['pink','magenta','hotpink','pink']
plt.stackplot(x,temperature, uvIndex, precipitation,colors=color ,labels=["Temperature", "UV Index", "Precipitation (%)"])

# titles and labels
plt.title('Stack Plot Example')
plt.xlabel('Index')
plt.ylabel('Values')



# Show plot
plt.legend(loc='upper left')
plt.show()
```

<h2> Violen Plots </h2>

What is Violin Plot?

A violin plot is a method of plotting numeric data and can be considered a combination of a box plot and a kernel density plot. It shows the distribution of the data across different categories and provides a visual summary of the data's density and probability distribution.

```python
#Data
y = [weather["Temperature"],weather["UV Index"],weather["Wind Speed"],weather["Precipitation (%)"]]

#Plotting 
plt.violinplot(y,showmeans=True,showmedians=True,showextrema=True,)

#Titles and labels

plt.title('Violin Plot Example')
plt.xlabel('Parameters')
plt.ylabel('Values')

plt.xticks([1,2,3,4],["Temperature", "UV Index", "Wind Speed", "Precipitation (%)"])
```
