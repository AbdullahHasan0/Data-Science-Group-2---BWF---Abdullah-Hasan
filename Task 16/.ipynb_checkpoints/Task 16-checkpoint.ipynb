<h1>Basic Plots:</h1>

<h3>Importing Libraries.</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h3>Importing Dataset.</h3>

```python
weatherData = pd.read_csv('weather.csv')
```

<h3> Exploring Data </h3>

```python
weather=weatherData.copy()
print(weather.columns,'\n')
print(weather.dtypes,'\n')

# Seperating Numeric Data
numericFeatures = weather.select_dtypes(include = "number")
print(numericFeatures)
```

<h2> Line Plot </h2>

What is Line Plot ?

A line plot, also known as a line chart or line graph, is a type of plot that displays data points connected by straight line segments. It is used to visualize how a single numerical variable or multiple numerical variables change over a continuous variable, such as time, index, or another ordered variable.

<h3>1. Create a simple line plot.</h3>

<p> I am making bar graph between total views on different platforms

```python
x1 = weather.index[:10]
y1 = weather["Atmospheric Pressure"][:10]

plt.xlabel("Starting 10 Rows")
plt.ylabel("Atmospheric Pressure in hPa")

#Plotting
plt.plot(x1,y1,color = 'g',marker = "o" , mfc = "blue",ms = 8)

#Showing the plot
plt.show()
```

<h3>2. Plot multiple lines on the same graph.</h3>

```python
# Select a range of data
x1 = weather.index[:10]
y1 = weather["Atmospheric Pressure"][:10]
y2 = weather["Temperature"][:10]
y3 = weather["Humidity"][:10]

# Making Figure Larger
plt.figure(figsize=(16, 12))  

# Plot multiple lines on the same plot
plt.plot(x1, y1, color='g', label='Atmospheric Pressure')
plt.plot(x1, y2, color='r',  label='Temperature')
plt.plot(x1, y3, color='y',  label='Humidity')

# Add titles and labels
plt.title('Comparison of Atmospheric Pressure, Temperature, and Humidity')
plt.xlabel('Index')
plt.ylabel('Values')

#Specifying scale on Y axis
plt.yticks(np.arange(0,1100,100))

# Showing the plot
plt.legend()
```

<h3>3. Add markers to a line plot.</h3>

```python
x1 = weather.index[:10]
y1 = weather["Atmospheric Pressure"][:10]
y2 = weather["Temperature"][:10]
y3 = weather["Humidity"][:10]

# Making Figure Larger
plt.figure(figsize=(16, 12))  

# Plot multiple lines on the same plot
plt.plot(x1, y1, color='g',marker = "o",mfc = "yellow",ms=8, label='Atmospheric Pressure')
plt.plot(x1, y2, color='r',marker = "s",mfc = "yellow",ms=8,  label='Temperature')
plt.plot(x1, y3, color='y',marker = "^",mfc = "yellow",ms=8,  label='Humidity')

# Add titles and labels
plt.title('Comparison of Atmospheric Pressure, Temperature, and Humidity')
plt.xlabel('Index')
plt.ylabel('Values')

#Specifying scale on Y axis
plt.yticks(np.arange(0,1100,100))


# Showing the plot
plt.legend()
```

<h3>4.Customize line styles and colors.</h3>

```python
x1 = weather.index[:10]
y1 = weather["Atmospheric Pressure"][:10]
y2 = weather["Temperature"][:10]
y3 = weather["Humidity"][:10]

# Making Figure Larger
plt.figure(figsize=(16, 12))

# Plotting
plt.plot(x1, y1, color='#355E3B',marker = "o",mfc = "yellow",ms=8,linestyle = "dotted", label='Atmospheric Pressure')
plt.plot(x1, y2, color='r',marker = "s",mfc = "yellow",ms=8,linestyle = "dashed",  label='Temperature')
plt.plot(x1, y3, color='y',marker = "^",mfc = "yellow",ms=8,linestyle = "dashdot",  label='Humidity')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Comparison of Atmospheric Pressure, Temperature, and Humidity')

#Specifying scale on Y axis
plt.yticks(np.arange(0,1100,100))

# Add a legend to differentiate the lines
plt.legend()
```

<h2>2. Scatter Plots</h2> 

What is Scatter Plot ?

A scatter plot visually represents data by plotting each observation as a point on a Cartesian plane. It uses two numerical variables where each point's position is determined by its x-coordinate (one variable) and y-coordinate (another variable). Scatter plots are valuable for assessing relationships or correlations between variables, offering insights into patterns, clusters, outliers, and the distribution of data points across both variables. They are widely used in data analysis and research to explore and visualize the interdependence and trends within datasets.

<h3>1. Create a basic scatter plot.</h3>

```python
x1 = weather["Precipitation (%)"][:50]
y1 = weather["Humidity"][:50]

#Plotting
plt.scatter(x1,y1)

#Printing Correlations
print(numericFeatures.corr())
```

<h4> Interpretation </h4>
Graph shows there is weak positive linear correlation between Humidity and Precipitation. This can be varified using above printed correlation table

<h3>2. Customize marker styles, sizes, and colors.</h3>

```python
x1 = weather["Precipitation (%)"][:50]
y1 = weather["Humidity"][:50]

#Plotting
plt.scatter(x1,y1,s = 30,marker = 's', alpha = 0.8,color='#FF6EC7',)

#Display the plot
plt.show()

```

<h3>3. Plot multiple scatter plots on the same graph.</h3>

```python
x1 = weather["Precipitation (%)"][:50]
y1 = weather["Humidity"][:50]
y2 = weather["Visibility (km)"][:50]

#Plotting
plt.scatter(x1,y1,s = 30,marker = 's', alpha = 0.8,color='#FF6EC7')
plt.scatter(x1,y2,s = 30,marker = 'v', alpha = 0.8,color='#006EC7')

#Display Data
plt.show()

```

<h3>4. Add labels and titles to the scatter plot.</h3>

```python
x1 = weather["Precipitation (%)"][:50]
y1 = weather["Humidity"][:50]
y2 = weather["Visibility (km)"][:50]

#Plotting 
plt.scatter(x1,y1,s = 30,marker = 's', alpha = 0.8,color='#FF6EC7',label = "Humidity")
plt.scatter(x1,y2,s = 30,marker = 'v', alpha = 0.8,color='#006EC7',label = "Visibility (km)")

#Giving Label And Title
plt.xlabel("Precipitation (%)")
plt.ylabel("Humidity & Visibility (km)")
plt.title ("Scatter Plot")

# Specifying Scale on Y Axis
plt.yticks(np.arange(0,110,5))

# Display the plot
plt.legend()
plt.show()

```

<h2>3. Bar Charts</h2> 

What is Bar Charts ?  
A bar chart is a type of graph that represents categorical data with rectangular bars. The lengths of these bars are proportional to the values they represent. Bar charts are commonly used for visualizing and comparing the frequencies, counts, or other numerical quantities associated with different categories or groups of data.

<h3>1. Create a vertical bar chart.</h3>

```python
# Calculate the mean temperature for each season
season_temperature = weather.groupby("Season")["Temperature"].mean()

# Extract the seasons and their corresponding mean temperatures
x = season_temperature.index
y = season_temperature.values

# Specifying Colors Of Bars
colors = ['green','yellow','orange','skyblue']

# Plotting the bar chart
plt.figure(figsize=(5, 10))
plt.bar(x, y, color=colors,width = 0.4)
plt.xlabel('Season')
plt.ylabel('Average Temperature')
plt.title('Average Temperature by Season')

# Display the plot
plt.show()
```

<h3>2. Create a horizontal bar chart.</h3>

```python
# Calculate the mean temperature for each season
season_temperature = weather.groupby("Season")["Temperature"].mean()

# Extract the seasons and their corresponding mean temperatures
x = season_temperature.index
y = season_temperature.values

# Specifying Colors Of Bars
colors = ['green','yellow','orange','skyblue']

# Plotting the bar chart
plt.figure(figsize=(5, 10))
plt.barh(x, y, color=colors,height=0.5)
plt.ylabel('Season')
plt.xlabel('Average Temperature')
plt.title('Average Temperature by Season')

# Display the plot
plt.show()
```

<h3>3. Customize bar colors and add edge colors.<h3>

```python
# Calculate the mean temperature for each season
season_temperature = weather.groupby("Season")["Temperature"].mean()

# Extract the seasons and their corresponding mean temperatures
x = season_temperature.index
y = season_temperature.values

# Specifying Colors Of Bars
colors = ['green','yellow','orange','skyblue']

# Plotting the bar chart
plt.figure(figsize=(5, 5))
plt.bar(x, y, color=colors,width = 0.4, edgecolor = 'grey',linewidth = 3)
plt.xlabel('Season')
plt.ylabel('Average Temperature')
plt.title('Average Temperature by Season')

# Display the plot
plt.show()
```

<h2>4. Pie Charts</h2> 

What is Pie Chart ?  
A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice represents a category of data, and the size of each slice is proportional to the quantity it represents. Pie charts are commonly used to show the composition or distribution of a dataset.

<h3>1. Create a basic pie chart.</h3>

```python
#Finding the dataset distribution of data over season

winterSeason = weather[weather["Season"]=='Winter']
winterSeasonCount = len(winterSeason)

summerSeason = weather[weather["Season"]=='Summer']
summerSeasonCount = len(summerSeason)

autumnSeason = weather[weather["Season"]=='Autumn']
autumnSeasonCount = len(autumnSeason)

springSeason = weather[weather["Season"]=='Spring']
springSeasonCount = len(springSeason)

totalCount = winterSeasonCount + summerSeasonCount + autumnSeasonCount + springSeasonCount

# Finding Percentages
winterPercentage = (winterSeasonCount/totalCount) * 100
summerPercentage = (summerSeasonCount/totalCount) * 100
autumnPercentage = (autumnSeasonCount/totalCount) * 100
springPercentage = (springSeasonCount/totalCount) * 100

# Plotting Pie Chart
y = np.array([winterPercentage,summerPercentage,autumnPercentage,springPercentage])
label = ["Winter","Summer","Autumn","Spring"]

plt.pie(y,labels=label)

```

<h3>2. Customize slice colors and add a legend.

```python
plt.pie(y,labels=label, colors = ['skyblue','yellow','green','orange'],shadow = True)
plt.legend()
plt.show()
```

<h3>3. Explode a slice of the pie chart. </h3> 

```python

plt.pie(y,labels=label, colors = ['skyblue','yellow','green','orange'],shadow = True,explode = [0.01,0.10,0.01,0.01],)

```

<h3> 4. Add percentage labels to slices

```python
plt.pie(y,labels=label, colors = ['skyblue','yellow','green','orange'],shadow = True,explode = [0.01,0.10,0.01,0.01],autopct = '%1.2f%%')
```
